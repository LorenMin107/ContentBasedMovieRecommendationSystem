```python
data.rename(columns={'Release Year': 'release_year'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
data.release_year.value_counts(dropna=False).sort_index().plot(kind='barh',figsize=(15,16))
plt.show()
```

```python
data.columns = data.columns.map(str.lower)
```

```python
data.columns
```

```python
data = data.loc[:,['directors','stars','genres','title']]
```

```python
data.head(10)
```

```python
data['stars'] = data['stars'].replace(np.nan, 'unknown')
data['directors'] = data['directors'].replace(np.nan, 'unknown')
```

```python
data
```

```python
data['genres'] = data['genres'].str.replace('|', ' ')
```

```python
data
```

```python
data['title'] = data['title'].str.lower()
```
